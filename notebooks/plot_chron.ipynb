```python
fig, ax = plt.subplots(1,1,figsize=(8,6), dpi=600)

# Step 1–2: Collect (x, y) pairs from pure_chiral domains
categories = ['Type I','Type II','Type II → Type I','Type II → Type III','rotates','stripe']
# energies: 'Edens_Zeeman', 'Edens_anis', 'Edens_demag', 'Edens_exch', 'Edens_total'
collect_edens = 'Edens_anis_core'
scale_to_geometry = True
aspect_ratio_thresh = 4

if collect_edens.endswith('_dw'):
    region = '_dw'
    energy_scale = 'surface_area'
elif collect_edens.endswith('_core'):
    region = '_core'
    energy_scale = 'volume'
else:
    region = ''
    energy_scale = 'volume_dilated'

for category in categories:
    frame_edens_pairs = []
    for domain_id, domain in domains_chron.items():
        if category == 'Type II':
            if category in domain.tags and 'rotates' not in domain.tags:
                for i, (frame_idx, energy, scale) in enumerate(zip(domain.frames, domain.energy[collect_edens], domain.properties[energy_scale])):
                    if scale_to_geometry:
                        energy = energy/scale
                    # Get aspect ratio (z) from the domain's proj dictionary
                    aspect_ratio = domain.proj['properties'][i]['aspect_ratio']
                    if aspect_ratio <= aspect_ratio_thresh and len(domain.frames) > 1:
                        frame_edens_pairs.append((frame_idx, energy))
        elif category == 'rotates':
            if category in domain.tags and 'Type II → Type III' not in domain.tags:
                for i, (frame_idx, energy, scale) in enumerate(zip(domain.frames, domain.energy[collect_edens], domain.properties[energy_scale])):
                    if scale_to_geometry:
                        energy = energy/scale
                    # Get aspect ratio (z) from the domain's proj dictionary
                    aspect_ratio = domain.proj['properties'][i]['aspect_ratio']
                    if aspect_ratio <= aspect_ratio_thresh and len(domain.frames) > 1:
                        frame_edens_pairs.append((frame_idx, energy))
        elif category == 'stripes':
            for i, (frame_idx, energy, scale) in enumerate(zip(domain.frames, domain.energy[collect_edens], domain.properties[energy_scale])):
                if scale_to_geometry:
                    energy = energy/scale
                frame_edens_pairs.append((frame_idx, energy))
        elif category in domain.tags:
            for i, (frame_idx, energy, scale) in enumerate(zip(domain.frames, domain.energy[collect_edens], domain.properties[energy_scale])):
                if scale_to_geometry:
                    if scale == 0:
                        continue
                    energy = energy/scale
                # Get aspect ratio (z) from the domain's proj dictionary
                aspect_ratio = domain.proj['properties'][i]['aspect_ratio']
                if aspect_ratio <= aspect_ratio_thresh and len(domain.frames) > 1:
                    frame_edens_pairs.append((frame_idx, energy))
    if category == 'rotates':
        category = 'Type II, rotates'

    # Step 3: Group y-values by x
    grouped = defaultdict(list)
    for frame, edens in frame_edens_pairs:
        grouped[frame].append(edens)

    # Step 4: Compute averages and std error
    averaged = {}
    errors = {}

    for frame, edenses in grouped.items():
        arr = np.array(edenses)
        averaged[frame] = np.mean(arr)
        errors[frame] = np.std(arr, ddof=1) / np.sqrt(len(arr)) if len(arr) > 1 else 0

    # Step 5: Sort and plot
    xs_sorted = np.array(sorted(averaged.keys()))
    Hz_i = 200
    dHz_frame = 0.25
    Hz_sorted = Hz_i + xs_sorted*dHz_frame
    ys_sorted = np.array([averaged[x] for x in xs_sorted])
    errs_sorted = np.array([errors[x] for x in xs_sorted])

    #plt.plot(xs_sorted, ys_sorted, marker='o', label=category)
    #plt.plot(xs_sorted, ys_sorted, label=category)
    #plt.scatter(xs_sorted, ys_sorted, label=category)
    ax.plot(Hz_sorted, ys_sorted, label=category)
    ax.fill_between(Hz_sorted,
                    ys_sorted - errs_sorted,
                    ys_sorted + errs_sorted,
                    alpha=0.3,)
    ax.set_xlabel("External Magnetic Field (mT)")
    ax.set_ylabel("Average Energy (J/m3)")
    ax.set_title(f"{collect_edens}")
    #ax.set_title("Average Soliton Energy by Category")

ax.legend()
ax.grid(True)
plt.show()
```
